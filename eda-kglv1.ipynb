```python
# fill -1 with probability in city_id and state_id
def fillneg1(col):
    null_vals = transactions[transactions[col] == -1]
    id_list = null_vals['card_id'].unique()
    all_null = []
    for cid in id_list:
        # print(cid)
        data = transactions[transactions['card_id'] == cid]
        prob_table = data[col].value_counts().to_frame()
        not_null = prob_table[prob_table.index != -1]
        if not_null.shape[0] == 0:
            all_null.append(cid)
            continue
        not_null = not_null / not_null.sum()
        null_count = prob_table.loc[-1]
        fill_vals = np.random.choice(not_null.index, null_count[0], p=not_null[col])
        transactions.loc[(transactions.card_id == cid) & (transactions.city_id == -1), col] = fill_vals
```
